```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data_n_0 = np.loadtxt('/Users/riyakore/Desktop/Astronomy/Astron_310/polytrope_n_0.txt')
z_n_0 = data_n_0[:,0]
w_n_0 = data_n_0[:,1]
dw_dz_n_0 = data_n_0[:,2]

z_n_0 = z_n_0/2.4495
```

```python
plt.plot(z_n_0, w_n_0, ls='--', label = 'numerical' , color = 'black')
plt.xlabel('z/zs')
plt.ylabel('w')
plt.title('For n = 0 polytrope')
plt.legend()
plt.show()
```

```python
# creates an array of range (0,1) containing 1000 elements
x_n_0 = np.linspace(0,1,1000)
x_n_0 = (data_n_0[:,0])
y_n_0 = 1 - ((x_n_0**2)/6)
x_n_0 = x_n_0 / 2.4495
plt.plot(x_n_0, y_n_0 , label = 'analytical', color = 'cyan')
plt.xlabel('z/zs')
plt.ylabel('w')
plt.title('For n = 0 polytrope')
plt.show()
```

```python
plt.plot(x_n_0 ,y_n_0 , label = 'analytical', color = 'cyan')
plt.plot(z_n_0, w_n_0, ls='--', label = 'numerical' , color = 'black')
plt.title('For n = 0 polytrope')
plt.legend()
plt.show()
```
