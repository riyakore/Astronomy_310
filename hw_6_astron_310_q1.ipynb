```python
data = pd.read_csv('/Users/riyakore/Desktop/Astronomy/Astron_310/MESA-Web_Job_10292320912/trimmed_history.data', skiprows=6, 
                   delim_whitespace=True, usecols = [1,3], names=["star_age", "log_L"])
print(data)
```

```python
data["log_star_age"] = np.log10(data["star_age"])
plt.figure(figsize=(10, 8))
plt.plot(data["log_star_age"], data["log_L"])
plt.xlabel("Log(t/yr)")
plt.ylabel("Log(L/L_0)")
plt.title("Luminosity(in solar luminosites) as a function of the star's age (in years)")
plt.grid(True)
plt.show()
```

```python
# Specify the value of x (log_star_age) you want to find
x_value_to_find = 7.10

# Find the closest x-value in the data
closest_x_value = data.iloc[(data["log_star_age"] - x_value_to_find).abs().idxmin()]["log_star_age"]

# Find the corresponding y (log_L) value
y_value = data[data["log_star_age"] == closest_x_value]["log_L"].values[0]

print(f"When log_star_age is closest to {x_value_to_find}, log_L is {y_value}")
```
