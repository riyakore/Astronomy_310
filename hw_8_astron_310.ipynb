```python
data_n_3 = np.loadtxt('/Users/riyakore/Desktop/Astronomy/Astron_310/polytrope_n_3.txt')

z = data_n_3[:,0]
w = data_n_3[:,1]
dw_dz = data_n_3[:,2]

#we need to use equation 19.4 from the handout 

central_density = 85.6955196444
log_density = (3*(np.log10(w))) + np.log10(central_density)
r = np.divide(z,6.89685)

plt.figure(figsize=(10, 8))
plt.plot(r, log_density, label = "Polyweb n = 3", color = 'blue')
plt.grid(True)
plt.xlabel("Radius")
plt.ylabel("Log_rho")
plt.legend(loc = 'lower left')
plt.title("Density as a function of radius for n = 3 Poly-Web")
plt.show()
```
