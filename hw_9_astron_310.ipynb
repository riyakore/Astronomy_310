```python
data_point_two = pd.read_csv('/Users/riyakore/Desktop/Astronomy/Astron_310/MESA-Web_Job_hw_9_0.2_solar_masses/profile8.data' 
                             ,skiprows=6, delim_whitespace=True, usecols = [0, 28], names=["mass", "h1"])
data_point_two
```
